# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

> ### <b> "In hospitals/facilities with number of discharges < 100, mean excess readmission rate is 1.023 and 63% have excess readmission rate greater than 1"
                                                                

```python
# Define a function: calculates output
def discharge_compare(discharge_dict):
    discharge = [hospital_read_df['Excess Readmission Ratio'][i] for i in list(discharge_dict.keys())]
    discharge_mean = sum(discharge) / len(discharge)
    greater = len([d for d in discharge if d > 1]) / len(discharge)
    return discharge_mean, greater
```

```python
# Check mean excess readmission rate for discharges < 100
discharge_100_dict = {index: v for (index, v) in dict(clean_hospital_read_df['Number of Discharges'][81:-3]).items() if v < 100}
discharge_100_mean, greater_100 = discharge_compare(discharge_100_dict)

print('Mean Excess Readmission Rate (discharges < 100): ', discharge_100_mean)
print('Excess Readmission Rate > 1 ((discharges < 100)): ', greater_100)
```

> ### <b> "Hospitals/facilties with small capacity (< 300) should be required to demonstrate upgraded resource allocation for quality care to continue operation."

#### Examine stats for small capacity hospitals

```python
def proportion_excess(index, data):
    
    # mean readmission
    readmission_mean = data.mean()
    
    # excess readmission >1 proportion
    index_excess = [i for i in index if df['readmission'][i] > 1]
    readmission_excess = df['readmission'][index_excess]
    rate_excess = len(readmission_excess) / len(index)
    return readmission_mean, rate_excess
```

```python
# stats for discharge <300 hospitals

index_300 = [i for i in df.index if df['discharge'].loc[i] < 300] 
readmission_300 = df['readmission'][index_300]

mean_300, excess_300 = proportion_excess(index_300, readmission_300)

print('Mean Excess Readmission Rate (discharges < 300): ', mean_300)
print('Excess Readmission Rate > 1 (discharges < 300): ', excess_300)
```

#### Examine stats for larger capacity hospitals

```python
# stats for all other hospitals

index_other = [i for i in df.index if df['discharge'].loc[i] > 300]
readmission_other = df['readmission'][index_other]

mean_other, excess_other = proportion_excess(index_other, readmission_other)

print('Mean Excess Readmission Rate (discharges < 300): ', mean_other)
print('Excess Readmission Rate > 1 (discharges < 300): ', excess_other)
```

Compared with "large hospital" category, hospitals in the "small" category (discharges < 300) have a higher proportion of excess readmission rates ('>1') and overall readmission rate. 
<p>
Follow-up w/ significance test below.

#### Determine appropriate stats test

```python
stat_1, p_1 = stats.normaltest(df.loc[index_300].readmission.values)
stat_2, p_2 = stats.normaltest(df.loc[index_300].readmission.values)

print('Pvalue for normality (small hospitals): ', p_1)
print('Pvalue for normality (large hospitals): ', p_2)
```

Scipy's normaltest tests the null hypothesis that a sample comes from a normal distribution. Since the p-values much smaller < 0.01, we reject null and therefore can't say for certain that their distributions are normal. So we proceed w/ bootstrapping a null distribution.

#### Bootstrap t-test:  small vs large average readmission rate

H0: Small hospitals w/ discharge <300 do not have different proportion of excess readmission rates ('>1') from larger hospitals.
<br>
HA: Small hospitals w/ discharge <300 have lower proportion of excess readmission rates ('>1') than larger hospitals.

```python
# bootstrap replicate function

def bootstrap_replicate_1d(data, func):
    bs_sample = np.random.choice(data, len(data))
    return func(bs_sample)

# Function that draws multiple bootstrap replicates

def draw_bs_reps(data, func, size=10000):

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates
```

```python
# bootstrap mean readmission, using mean = larger hospital's mean readmission as null hypothesis

bs_replicates = draw_bs_reps(readmission_other, np.mean, size=10000)

p = np.sum( bs_replicates >= np.mean(readmission_300) ) / len(bs_replicates)
print('P-value: ', p)
```

P-value much smaller < 0.01, so reject null that there's no difference. Sufficient evidence that small hospitals indeed have lower average readmission rate.
<p>
Assessment that hospitals with discharges ('< 300') should upgrade quality care is backed by this evidence. However, should examine other parameters in data, and this result may change depending on how "small hospital" is categorized by discharge (i.e. from earlier recommendation that adjust criteria for small & large discharge counts).

> ### <b> "Directives and incentives should be provided for consolidation of hospitals and facilities to have a smaller number of them with higher capacity and number of discharges."

Although higher capacity (defined under current "large discharge" criteria) is correlated with lower readmission rates, there's likely many other contributing factors to hospitals' success at keeping readmission rates low (i.e. other parameters in the dataset). Furthermore, combining hospitals introduces many more confounding factors, including an impact on administration and hospital efficiency, which could actually increase readmission rates and reverse the trend.
<p>
<b>Recommendations</b>
- Analyze impact of other parameters besides hospital discharge count on readmission rate
- Map out other confounding factors that could affect readmission rates if smaller hospitals were to consolidate

## B: Discussion of results

<b>Weak trend</b>. Although the trend between discharge count (hospital capacity) vs. readmission rates is statistically significant, both the correlation (R value -0.09) and slope of this trend (-2.86e-05) are extremely weak. According to this trend, for instance, a hospital expansion of at least ~3500 additional discharges must be implemented for every 0.1 decrease in readmission ratio.

<b>Costly changes</b>. These costly changes call into question the practical significance of implementing the recommendations called for in the report, no matter how statistically significant they may be. Unless there is sufficient funding to undertake such a large task, and not taking into consideration the additional effects it would have on hospital administration and efficiency, the report may be not be of practical significance.

<b>Possible solution</b>. A middle ground is to re-examine the report's criteria for small & large discharge hospitals. 
- As our scatterplot examination showed, there's actually quite a large cluster of data that is not included in either criteria. Leaving out this data may be a reason that our cross-check of the claim that smaller hospitals having greater incidence of excess readmissions turned out to be statistically insignificant.
- Furthermore, redefining criteria for hospital size, and in addition introducing the effects of other parameters, could also specify which hospitals for which costly policy changes could be worthwhile.

<b>Check other parameters</b>. Finally, future reports should include an examination of not just discharge vs. readmission, but also other valuable parameters left out of this report such as state or start & end dates:
- The hospital's state of residence, for example, could allow us to narrow results to specific states that disproportionately increase the mean readmission rate because of their unusually high excess readmissions. Such a parameter would be especially relevant if the report is intended to recommend state-wide policy changes.
- Start & end dates could determine whether a policy change is still relevant, if the excess readmission rates happened far in the past.

> #### Construct another plot that conveys the same information in a more direct manner.

A violin plot visualizes data density for univariate data and curved distribution (KDE) approximating the data's histogram. Since the data already effectively comprise three groups: low discharge, high discharge, and uncategorized data, we can use category as the univariable.

#### Categorize data for plot

```python
# Organize categorical data

index_low = [i for i in df.index if df['discharge'].loc[i] <= 350] 
index_low = [i for i in df.index if df['readmission'].loc[i] <= 2 and df['readmission'].loc[i] >= 1.15]

index_high = [i for i in df.index if df['discharge'].loc[i] <= 2500 and df['discharge'].loc[i] >= 800] 
index_high = [i for i in df.index if df['readmission'].loc[i] <= .95 and df['discharge'].loc[i] >= .5]

index_other = [i for i in df.index if i not in index_low + index_high]

# Add category info to df

df['category'] = np.nan # placeholder
df.loc[index_low, 'category'] = 'low'
df.loc[index_high, 'category'] = 'high'
df.loc[index_other, 'category'] = 'other'

df.head()
```

```python
fig, ax = plt.subplots(figsize=(14,9))

# violin plot
sns.violinplot(ax=ax, x='category', y='readmission', data=df, palette='Greys');

# overlay strip plot w/ jitter
sns.stripplot(ax=ax, x='category', y='readmission', data=df, size=5, jitter=True, alpha=.2);
```
